```python
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
from libs import Hough
import math
import cv2
```

```python
image = cv2.imread("./images/Line.png")
```

```python
grayscale = Hough.grayImg(image)
edges = Hough.canny_edge_detection(grayscale)
```

```python
plt.subplot(131), plt.imshow(image)
accumulator =Hough.hough_line(edges,2)
Hough.superimpose(image,accumulator,110)
```
